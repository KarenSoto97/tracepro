```python
from tracepro.app import TP_app
from tracepro.element import Element
from tracepro.source import Source
```

### New file

```python
working_path =r"C:\Software\tracepro\examples\results\rotation"
tp_file_name = "rotation_example"

tp = TP_app(macro_path=working_path, macro_name='rotate')
tp.new_file(file_name=tp_file_name, tp_path=working_path)
```

### Block

```python
bloque = Element(name="block")
block = bloque.block(dimensions=(1, 1, 0.5), center=(0, 5, 5), orientation_method=("angles", (90, 0, 0), ""), degrees=True)
tp.add_function(block)
tp.name_surfaces(name="Cuadrado", object_name=bloque.name, surface_num=0)
# tp.execute()
```

### Rotation

```python
angle = -45
rotation_point = (0, 0, 0)

tp.rotate_object(object_name=bloque.name, angle=angle, axis_type=4, degrees=True, rotation_point=rotation_point, object_ref=True)
tp.save_file(file_name=tp_file_name, tp_path=working_path)
result = tp.execute(graphic_window=True)
```
