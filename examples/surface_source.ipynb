```python
from tracepro.app import TP_app
from tracepro.element import Element
from tracepro.source import Source
```

# Surface source

### New file

```python
working_path =r"C:\Software\tracepro\examples\results\surface_source"
tp_file_name = "surface_source_example"

tp = TP_app(macro_path=working_path, macro_name="surface_source")
tp.new_file(file_name=tp_file_name, tp_path=working_path)
```

### Emisor

```python
dimensions = (1, 1, 1) # x, y, z
center_pos = (0, 0, 0)
rotation = (0, 0, 0)
orientation_method = ("angles", rotation, "")
degrees = False
```

```python
emisor = Element(name="emisor")
bloque = emisor.block(dimensions=dimensions, center=center_pos, orientation_method=orientation_method, degrees=degrees)
tp.add_function(bloque)

fuente = Source(name="fuente", source_num=0)
surface_source = fuente.add_surface_source(object_name=emisor.name, surface_num=1, emission_type="irradiance", units="radiometric", 
                                           emission=3, ang_dist=2, flux=1000)
tp.add_function(surface_source)
# tp.name_surfaces(object_name=emisor.name, surface_num=1, name="Plano de emision")
# fuente.source_ID = f"{emisor.id}/Plano de emision"

# NOTE:
# In TracePro, renaming a surface after assigning it as a surface source
# breaks the internal source reference. Therefore, surface names should
# not be modified once a surface source has been created.

wavelength = fuente.set_wavelength(wavelength=(0.555, 1))
tp.add_function(wavelength)
```

### Detector

```python
detector = Element(name="detector")
detec = detector.block(dimensions=dimensions, center=(0, 0, 5), orientation_method=orientation_method, degrees=degrees)
tp.add_function(detec)
tp.apply_property(object_name=detector.name, surface_num=0, property=("default", "perfect absorber"))
tp.name_surfaces(object_name=detector.name, surface_num=0, name="Plano deteccion")
```

```python
raytrace = Source.raytrace()
tp.add_function(raytrace)

tp.select_detection_surface(object_name=detector.name, surface_num=0)
irradiance_map = fuente.irradiance_analysis(ray_type="incident")
tp.add_function(irradiance_map)
tp.save_file(file_name=tp_file_name, tp_path=working_path)
tp.execute(graphic_window=True)
```
